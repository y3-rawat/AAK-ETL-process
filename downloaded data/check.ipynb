```python
import requests
import io
import pandas as pd

# URL for the World Bank API data for Afghanistan
url = "https://api.worldbank.org/v2/en/country/AFG?downloadformat=csv"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Read the content of the response
    content = response.content
    
    # Use StringIO to create a file-like object from the content
    csv_file = io.StringIO(content.decode('utf-8'))
    
    # Read the CSV data into a pandas DataFrame
    df_afghanistan = pd.read_csv(csv_file)
    
    # Display the first few rows of the dataframe
    print(df_afghanistan.head())
else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")
```
