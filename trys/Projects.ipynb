```python
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import pandas as pd

# Ensure dates are in datetime format
df['start_date'] = pd.to_datetime(df['start_date'], errors='coerce')
df['end_date'] = pd.to_datetime(df['end_date'], errors='coerce')

# 1. Project Distribution by Country
plt.figure(figsize=(12, 6))
country_counts = df['countryname'].value_counts()
country_counts.plot(kind='bar')
plt.title('Project Distribution by Country')
plt.xlabel('Country')
plt.ylabel('Number of Projects')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 2. Project Timeline
plt.figure(figsize=(12, 6))
valid_dates = df.dropna(subset=['start_date', 'end_date'])
plt.scatter(valid_dates['start_date'], valid_dates['end_date'])
plt.title('Project Timeline')
plt.xlabel('Start Date')
plt.ylabel('End Date')
plt.tight_layout()
plt.show()

# 3. Project Status Distribution
plt.figure(figsize=(8, 8))
status_counts = df['disclosure_status'].value_counts()
status_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Project Status Distribution')
plt.ylabel('')
plt.tight_layout()
plt.show()

# 4. Project Counts Over Time
yearly_counts = df['start_date'].dt.year.value_counts().sort_index()
plt.figure(figsize=(12, 6))
yearly_counts.plot(kind='line')
plt.title('Number of Projects Started Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Projects')
plt.tight_layout()
plt.show()

# 5. Project Duration
df['duration'] = (df['end_date'] - df['start_date']).dt.days
plt.figure(figsize=(12, 6))
sns.histplot(data=df.dropna(subset=['duration']), x='duration', bins=20)
plt.title('Distribution of Project Durations')
plt.xlabel('Duration (days)')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# 6. Word Cloud
text = ' '.join(df['project_name'].dropna())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Project Names')
plt.tight_layout()
plt.show()
```
